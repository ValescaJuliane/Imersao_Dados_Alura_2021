# Desafio 5: Escolha a droga mais utilizada e tente criar um modelo para prever está label

Para este desafio vamos classificar os dois compostos mais utilizados

```python
#Identificando os dois compostos mais utilizado, onde 'cacb2b860' é o controle

dados_combinados['composto'].value_counts().head(3)
```

```python
#Selecionando as linhas na base de dados onde os compostos que iremos analizar aparecem

compostos_analise = ['87d714366', '9f80f3f77']
dados_analise = dados_combinados.query('composto in @compostos_analise')

dados_analise
```

```python
#Utilizando o algoritmo SVM, ou máquina de vetor de suporte, com kernel linear

from sklearn import svm

x = dados_analise.drop(['id','composto','tratamento'], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_analise['composto']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_svm = svm.NuSVC(kernel='linear')
modelo_svm.fit(x_treino, y_treino)
teste = (modelo_svm.score(x_teste, y_teste))
treino = (modelo_svm.score(x_treino, y_treino))
```

```python
treino
```

```python
teste
```

Para esta análise binária, para classificar se os compostos utilizados em cada experimento é o c'87d714366' ou '9f80f3f77', o algoritmos SVM conseguiu 100% de score
