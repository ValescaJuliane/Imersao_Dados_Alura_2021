# Desafio 1: Criar tabelas de frequência utilizando o pandas.groupby()

O pandas.groupby também é uma forma de agrupamento de dados, podendo ter a aplicação de funções e combinação de resultados.


Abaixo vemos a frequência dados relacionados ao tempo agrupados pelas doses

```python
dados.groupby(by = ['dose'])['tempo'].value_counts().unstack()
```

Abaixo temos a frequência dos dados de tratamento agrupados pela dose e pelo tempo de exposição

```python
dados.groupby(by = [dados['dose'], dados['tempo']])['tratamento'].value_counts().unstack()
```

Abaixo temos a frêquenca média dos dados do gene g-0 agrupados por dose, tempo e o tipo de tratamento aplicado

```python
dados.groupby(by = [dados['dose'], dados['tempo'], dados['tratamento']])['g-0'].mean()
```

# Desafio 2: Normalizar o crosstab pela coluna

Na função crosstab é possível normalizar os dados das colunas setando o atributo normalize = 'columns'

```python
pd.crosstab([dados['dose'], dados['tempo']],  dados['tratamento'], normalize='columns')
```

# Desafio 3: Eplorar outros agregadores na frunção crosstab

O tipo de agregador 'var' é a variância.

Dado um conjunto de dados, a variância é uma medida de dispersão que mostra o quão distante cada valor desse conjunto está do valor central (médio).

Quanto menor é a variância, mais próximos os valores estão da média; mas quanto maior ela é, mais os valores estão distantes da média.

```python
pd.crosstab([dados['dose'], dados['tempo']],  dados['tratamento'], values=dados['g-0'], aggfunc='var')
```

# Desafio 4: Explorar o melt

A função Pandas melt () é usada para alterar o formato do DataFrame. É usado para criar um formato específico do objeto DataFrame onde uma ou mais colunas funcionam como identificadores. Todas as colunas restantes são tratadas como valores.

```python
pd.melt(dados, id_vars= ['dose', 'tratamento'], value_vars= ['tempo'] )
```

# Desafio 5: Calcular e analisar a correlação entre G e C. Refletir sobre os efeitos biológicos.

Correlação entre os 50 primeiro Gs e os 50 primeiros Cs

```python
#A variável genes recebe as colunas de g-0 até g-50
genes = dados.loc[:,'g-0':'g-50']

#A variável celulas recebe as colunas de c-0 até c-50
celulas = dados.loc[:,'c-0':'c-50']

#A variável tab_GeC será uma tabela resultante da concatenação de genes e celulas
tab_GeC = pd.concat([genes, celulas], axis=1)
tab_GeC
```

```python
# Agora podemos fazer a correlação entre os Gs e os Cs
# Onde as linhas vão de c-0 a c-50 e as colunas de g-0 a g-50

correlacao = tab_GeC.corr().loc['c-0':'c-50','g-0':'g-50']
correlacao
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(correlacao, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlacao, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Desafio 6: Estudar o código que plota a matriz de correlação (heatmap)

Uma matriz de correlação é simétrica, entrão primeiramente é feita uma mask com o numpy para deixar apenas o triângulo de baixo da matriz. Com o np.ones_like o tipo de resultado é definido como um boolean. Com o plt.subplots é definido um tamanho para o gráfico que vai ser plotado, e com o sns.diverging_palette é definido as cores e saturações para o heatmap, finalmente, com o sns.heatmap, o gráfico é plotado.

```python
mask = np.triu(np.ones_like(correlacao, dtype=bool))

f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(correlacao, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
