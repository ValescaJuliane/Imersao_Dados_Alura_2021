# Desafio 02: Criar a coluna eh_controle para quando na linha tratamento == com_controle

```python
#Inserindo os dados na coluna 'eh_controle' 
#Temos um for que vai do índice 0 até o tamanho da tabela
#Para cada índice da coluna tratamento é verificado se é com_controle
#Se sim, a coluna 'eh_controle' recebe 1, se não, recebe 0

for i in range(0,len(dados_combinados)):
    if dados_combinados.loc[i,'tratamento']=='com_controle':
        dados_combinados.loc[i,'eh_controle'] = 1
    else:
        dados_combinados.loc[i,'eh_controle'] = 0
```

# Desafio 7: Descobrir se tem algum composto que dependendo da configuração do experimento, ativa MOAs diferentes

```python
#Para esse desafio precisamos unir todos os dados

dados_completos = pd.merge(dados, dados_resultados.loc[:,:'wnt_inhibitor'], on='id')
dados_completos.head()
```

```python
#Criando a lista com todos os MoA's

moas_completo = dados_resultados.drop('id',axis=1).columns[:-2]
moas_completo
```

```python
#Criando uma tabela agrupado por composto
dados_agrupados = dados_completos.groupby(by='composto')
```

```python
#Utilizar o loop para verificar

for moa in moas_completo:
    count = 0
    grupo = dados_agrupados[moa].nunique() #Calcula os valores unicos dos MoA's em questão para cada composto analisado
    for i in range(len(grupo)):
        if grupo[i] == 1:
            pass
        else:
            count += 1
            print(grupo[i].index)
            
    if count == 1:
        print(f'MoA {moa} é ativado ou não dependendo das condições')
print('Fim')
```

Como count nunca recebe o valor 1, vemos que nenhum MoA diferente foi ativado dependendo das condições.
